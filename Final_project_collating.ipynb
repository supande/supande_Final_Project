```python
import csv,datetime

f=open("Crime_complaints_2020_2.csv",'r')
reader=csv.reader(f)
f2=open("NumberOfCrimesPerDay_2020_2.csv",'w')
writer=csv.writer(f2,delimiter=',',lineterminator='\n')
d={}
n=0
for row in reader:
    if n>0:
        date=row[1]
        if date in d.keys():
            d[date]+=1
        else:
            d[date]=1
    n+=1
    
#print(d)

l1=["Date","Number of crimes"]
writer.writerow(l1)

for k,v in sorted(d.items()):
    l=[k,v]
    #print(l)
    writer.writerow(l)
    
f.close()
f2.close()
    
```

```python
import csv,datetime

f=open("NYPD_Complaint_Data_Current__Year_To_Date_ (1).csv",'r')
reader=csv.reader(f)
f2=open("NumberOfCrimesPerDay_2021.csv",'w')
writer=csv.writer(f2,delimiter=',',lineterminator='\n')
d={}
n=0
for row in reader:
    if n>0:
        date=None
        try:
            date=row[3]
        except Exception as e:
            continue
        if date is not None:
            dto=datetime.datetime.strptime(date,"%m/%d/%Y")
            if dto>datetime.datetime.strptime('12/31/2020',"%m/%d/%Y") and dto<datetime.datetime.strptime('10/01/2021',"%m/%d/%Y"):
                if date in d.keys():
                    d[date]+=1
                else:
                    d[date]=1
    n+=1
#print(d)   
#print(d)

l1=["Date","Number of crimes"]
writer.writerow(l1)
l=[]
for k,v in sorted(d.items()):
    l=[k,v]
    #print(l)
    writer.writerow(l)
f.close()
f2.close()
```
