```python
import csv,datetime

f=open("NYPD_Complaint_Data_Historic.csv",'r')
reader=csv.reader(f)

f2=open("Crime_complaints_2019.csv",'w')
writer=csv.writer(f2,delimiter=',',lineterminator='\n')

n=0

for row in reader:
    if n==0:
        writer.writerow(row)
    if n>0:
        date=None
        try:
            date=row[1]
        except:
            continue
        if date is not None:
             
            dto=None
            try:
                dto=datetime.datetime.strptime(date,"%m/%d/%Y")
            except:
                continue
            if dto is not None:
                if dto>datetime.datetime.strptime('12/31/2018',"%m/%d/%Y") and dto<datetime.datetime.strptime('01/01/2020',"%m/%d/%Y"):
                    writer.writerow(row)
    
    n+=1
    

f.close()
f2.close()
        
```

```python
import csv,datetime

f=open("COVID-19_Daily_Counts_of_Cases__Hospitalizations__and_Deaths.csv",'r')
reader=csv.reader(f)

f2=open("CovidDailyCases_2020.csv",'w')
writer=csv.writer(f2,delimiter=',',lineterminator='\n')

n=0

for row in reader:
    l=[]
    l.append(row[1])
    l.append(row[3])
    l.append(row[4])
    l.append(row[6])
    
    if n==0:
        writer.writerow(l)
    if n>0:
        dto=datetime.datetime.strptime(row[0],"%m/%d/%Y")
        if dto>datetime.datetime.strptime('02/29/2020',"%m/%d/%Y") and dto<datetime.datetime.strptime('01/01/2021',"%m/%d/%Y"):
            writer.writerow(l)
    n+=1
    

f.close()
f2.close()
```
